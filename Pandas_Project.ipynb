## Data Cleaning
The four features that we will be working with most frequently henceforth are Installs, Size, Rating and Price. The info() function (from the previous task) told us that Installs and Price columns are of type object and not int64 or float64 as we would expect. This is because the column contains some characters more than just [0,9] digits. Ideally, we would want these columns to be numeric as their name suggests.
Hence, we now proceed to data cleaning and prepare our data to be consumed in our analyis later. Specifically, the presence of special characters (, $ +) in the Installs and Price columns make their conversion to a numerical data type difficult

```python
df['Installs']=df['Installs'].astype(str).str.replace('[^0-9]','')
```

```python
df['Installs']=pd.to_numeric(df['Installs'])
```

```python
df['Price']=df['Price'].astype(str).str.replace('[^0-9]','')
```

```python
df['Price']=pd.to_numeric(df['Price'])
```

2. Fill the null values in Rating column depending on the Category column

```python
df.head()
```

```python
grpc=df.groupby('Category')
```

```python
for i in df['Category'].unique():
    df['Rating'][df['Category']==i]=df['Rating'][df['Category']==i].fillna(grpc['Rating'].mean()[i])
```

3. Create a binned column using the Rating column which should contain 3 bins namely less than 3, 3 to 4, 4 to 5

```python
df['Rating_bin']=pd.cut(df['Rating'],bins=(0,3,4,5),labels=('0 t0 less than 3','3 to 4','4 to 5'))
```

5. How many apps based on content rating which Size are less than 10.

```python
no_of_app=df[df['Size']<10].groupby('Content Rating')['App'].count()
print('no of app based on content rating are less than 10: ',no_of_app)
```

```python
total_no_app=no_of_app.sum()
print('Total no of app based on content rating are less than 10: ',total_no_app)
```
